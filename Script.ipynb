this project is to change the data from sperated columns of table in one excel into one table orgnized data

```python
data_import=pd.read_csv("data_cleaning_challenge.csv")
data_import
```

```python
drop_extra=data_import.drop(columns=["Unnamed: 9","Unnamed: 10"])
```

```python
drop_extra
```

```python
drop_nans=drop_extra[drop_extra["Row Type"].notna()]
drop_nans
```

```python
column_values=[]
counter=0
for i in drop_nans["Row Type"]:
    
    if("first name") in i:
        counter+=1
    column_values.append(counter)
   
```

```python
column_values
```

```python
iter_cols=drop_nans
iter_cols["Iteration"]=column_values
iter_cols
```

```python
drop_extra_column_row=iter_cols[iter_cols["Row Type"]!="Row Type"]
drop_extra_column_row
```

```python
#creatae a new table have only the rows of the first names (and will take the whole columns)
name_dataframe=drop_extra_column_row[drop_extra_column_row["Row Type"].str.contains("first name")]
#change the columns names because it coms from the previouos version and drop the extraa
name_dataframe.drop(columns=["Speed1","Speed2","Electricity","Effort","Weight","Torque"],inplace=True)
name_dataframe.rename(columns={"Row Type":"First Name",	"Iter Number":"Last Name","Power1":"Date"}, inplace=True)
#remove extra info tag on the data for the specefic columns "with the specefic extra tag fixed"
name_dataframe["First Name"]=name_dataframe["First Name"].str[12:]
name_dataframe["Last Name"]=name_dataframe["Last Name"].str[11:]
name_dataframe["Date"]=name_dataframe["Date"].str[6:]


name_dataframe
```

```python
no_name_dataframe = drop_extra_column_row[~drop_extra_column_row["Row Type"].str.contains(
    "first name")]
no_name_dataframe
```

```python
pd.merge(left=name_dataframe, right=no_name_dataframe, how="inner", on="Iteration")
```
